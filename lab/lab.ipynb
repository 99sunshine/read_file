```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.text import Text

import numpy

fig1 = plt.figure(facecolor='white')
ax1 = plt.axes(frameon=False)


#ax1.set_frame_on(False)     # Alternate way to turn frame off

ax1.get_xaxis().tick_bottom()   # Turn off ticks at top of plot

#ax1.axes.get_xaxis().set_visible(False)
#ax1.axes.get_yaxis().set_visible(False)     # Hide y axis

# Add a plot
y_offset = 2.0
x = numpy.arange(-5.0, 5.0, 0.1)
for i in range(3):
    ax1.plot(x, numpy.sin((i + 1) * x) + i * y_offset, label=str(i+1))
ax1.errorbar(x, x, yerr = x)

#ax2.plot(x, np.cos(x))

plt.legend(loc=0)

# Draw the x axis line
# Note that this must be done after plotting, to get the correct
# view interval
xmin, xmax = ax1.get_xaxis().get_view_interval()
ymin, ymax = ax1.get_yaxis().get_view_interval()
ax1.add_artist(Line2D((xmin, xmax), (ymin, ymin), color='black', linewidth=2))

plt.show()
```

```python
t = np.arange(0.0, 2.0, 0.01)

s1 = np.sin(2 * np.pi * t)
s2 = np.exp(-t)
s3 = s1 * s2

fig, axs = plt.subplots(3, 1, sharex=True)
# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)

# Plot each graph, and manually set the y tick values
axs[0].plot(t, s1)
axs[0].set_yticks(np.arange(-0.9, 1.0, 0.4))
axs[0].set_ylim(-1, 1)

axs[0].plot(t, s1*2)
axs[0].set_yticks(np.arange(-0.9, 1.0, 0.4))
axs[0].set_ylim(-1, 1)

axs[1].plot(t, s2)
axs[1].set_yticks(np.arange(0.1, 1.0, 0.2))
axs[1].set_ylim(0, 1)

axs[2].plot(t, s3)
axs[2].set_yticks(np.arange(-0.9, 1.0, 0.4))
axs[2].set_ylim(-1, 1)

#remove spine------------------------------------------------

axs[0].spines['bottom'].set_visible(False)
 
axs[1].spines['bottom'].set_visible(False) #yayyyyyyyyyyy
axs[1].spines['top'].set_visible(False) 
axs[2].spines['top'].set_visible(False)  

#move the spine
pos1 = axs[0].get_position() # get the original position 
pos2 = [pos1.x0, pos1.y0 + 0.03,  pos1.width, pos1.height] 
axs[0].set_position(pos2) # set a new position

pos3 = axs[1].get_position() # get the original position 
pos4 = [pos3.x0, pos3.y0 + 0.01,  pos3.width, pos3.height] 
axs[1].set_position(pos4) # set a new position

#grid---------------------------------------------------------

#axs[0].yaxis.grid(True)
axs[0].xaxis.grid(True)
#axs[1].yaxis.grid(True)
axs[1].xaxis.grid(True)
#axs[2].yaxis.grid(True)
axs[2].xaxis.grid(True)

#plt.xlabel('Solar Longitude', fontsize = 20)
#plt.ylabel('Volume Mixing Ratios', fontsize = 20)

plt.show()
```
